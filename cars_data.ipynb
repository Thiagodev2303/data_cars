```python
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Defining archive for use in database
car_df = pd.read_csv('used_cars_data.csv')
display(car_df.head(5))
```

```python
#Cleaning the data
car_df = car_df.dropna()    
```

```python
#Print only name and year of cars
name_year = car_df[['Name', 'Year']]
display(name_year)
```

```python
#Print firts 10 cars that were made between 2010 and 2019 from name_year data
from_2000_to2019 = name_year[(name_year['Year'] >= 2010) & (name_year['Year'] <= 2019)]
display(from_2000_to2019)
```

```python
#Print the quantity of cars per year
quantity_ofcars_per_year = from_2000_to2019['Year'].value_counts()
display(quantity_ofcars_per_year)
```

```python
#Create a graphic
line_graph = sns.lineplot(x=quantity_ofcars_per_year.index, y=quantity_ofcars_per_year.values)
line_graph.set_title('Quantity of cars per year')
plt.show(line_graph)
```
